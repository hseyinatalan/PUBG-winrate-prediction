# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

EDA için gerekli kütüphaneleri ekliyoruz.

```python
train = pd.read_csv("/kaggle/input/pubg-finish-placement-prediction/train_V2.csv")
test = pd.read_csv("/kaggle/input/pubg-finish-placement-prediction/test_V2.csv")
```

Verimizi yüklüyoruz.

```python
train.head()
train.info()
train.describe()
```

Veri yapısına bakıyoruz.İncelediğimde dikkatimi;
* "kills" durmuna bakınca çoğu oyuncu öldürme yapmamış veya 1-2 kill ile bitirmiş.
* "damageDealt" durumunda hasar dağılımı çok değişken, bazı oyuncular çok yüksek damage yapmış.
* "walkDistance" durumunda bazı oyuncular çok uzun mesafe yürümüş; outlier(aykırı değer) olabilir.
* "rideDistance" durumunda çoğu oyuncu araç kullanmamış, bazıları çok uzun mesafe sürmüş.
* "killPlace" durumunda oyuncuların çoğu orta sıralarda bitirmiş.

```python
train.isnull().sum()
```

Eksik değerlere bakıyoruz.Tek eksik değer, hedef değişken winPlacePerc sütununda 1 satır eksik.Sütunlarda eksik değer olmadığı için ekstra bir imputation (doldurma) işlemi gerekmiyor.

```python
sns.histplot(train['winPlacePerc'], bins=50, kde=True)
plt.show()
```

Hedef değişkeni inceliyorum.

winPlacePerc (0–1 arası bitirme yüzdesi)

En yüksek değer oyunun sonuncu bitirenlerin oranıdır.

```python
numeric_features = train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = train.select_dtypes(include=['object']).columns
```

Kategorik ve sayısal değişkenleri ayırıyorum.

```python
plt.figure(figsize=(12,10))
sns.heatmap(train[numeric_features].corr(), annot=True, fmt=".2f")
plt.show()
```

Korelasyon matrisinde dikkat çekenler;
* winPlacePerc ile yüksek pozitif korelasyon:

    killPlace (0.72)
    
    kills (0.42)
    
    damageDealt (0.44)
    
    walkDistance (0.81)
    
    weaponsAcquired (0.58)
    
    rideDistance (0.34)

→ Oyuncuların daha çok hareket etmesi (özellikle yürüyerek) ve daha iyi öldürme sıralamasına sahip olması başarıda en belirleyici faktörler.
* kills ile en yüksek korelasyonlar:

    killPlace (0.73)
    
    damageDealt (0.89)
→ Çok kill yapan oyuncular genellikle yüksek killPlace ve yüksek hasar değerlerine sahip.

```python
for col in ['kills', 'damageDealt', 'walkDistance', 'rideDistance']:
    plt.figure()
    sns.histplot(train[col], bins=50, kde=True)
    plt.title(col)
    plt.show()
```

Özelliklerin dağılımına ve outlier(aykırı değerlere)’lara bakalım.
* kills grafiğine bakınca çoğu oyuncu maçta az veya hiç kill yapmazken, az sayıda oyuncu yüksek kill yaparak başarıyı etkiliyor.
* damageDealt grafiğinde;

   * Çoğu oyuncunun maçta çok düşük hasar verdiği görülüyor, yani dağılım 0’a çok yakın ve sağa çarpık.
   * Hasar miktarı küçük değerlere çok yoğunlaşmış, bin veya daha üzerindeki hasar değerleri çok daha az sayıda oyuncuya ait.
   * Bu dağılım, oyuncuların çoğunun maçta sınırlı etki yaptığını, az sayıda oyuncunun yüksek hasar vererek başarıya katkıda bulunduğunu gösteriyor.

* walkDistance grafiğinde yüksek çoğunluk az yürüyor.
* rideDistance grafiğinde ise araç kullanımının çok düşün olduğu görülüyor.

```python
train = train.dropna(subset=['winPlacePerc'])
```
